```python
from gwpy.timeseries import TimeSeries

# Define GPS times
start = 1439800939.5
end   = 1439811771.375

# Fetch the auxiliary channel
ts = TimeSeries.fetch('H1:LSC-REFL_A_LF_OUT_DQ', start, end)

# Plot the time series
plot = ts.plot()
plot.gca().set_title("H1:LSC-REFL_A_LF_OUT_DQ Time Series")
plot.show()
```

```python



# Make a q-transform
q = ts.q_transform(outseg=(start, end))

# Plot the q-scan
plot = q.plot()
plot.gca().set_title("Q-scan of H1:LSC-REFL_A_LF_OUT_DQ")
plot.show()
```
