```python
from gwpy.timeseries import TimeSeries
import matplotlib.pyplot as plt

# glitch GPS times
g0, g1 = 1245062053.5, 1245062054.25
pad = 2
t0, t1 = g0 - pad, g1 + pad

# fetch strain
strain = TimeSeries.get("H1:DCS-CALIB_STRAIN_CLEAN_C01", t0, t1)

# Q-transform: specify fband using frange instead of fmin/fmax
qscan = strain.q_transform(
    frange=(10, 2048),   # frequency range in Hz
    qrange=(8, 64),      # Q values
    logf=True,           # logarithmic frequency axis
    whiten=True,         # whiten first for better visibility
    fftlength=0.25,      # shorter FFT for short glitch
    overlap=0.125
)

# plot
plot = qscan.plot()
plot.gca().axvspan(g0, g1, color="orange", alpha=0.3)
plt.title(f"H1 Q-transform around GPS {g0:.2f}-{g1:.2f}")
plt.show()
```
